Что-то сложное. Надо потом разобратся, по-сути нам нужны только имена файлов как одельная колнка. Названия имен будем брать напрямую из файлов. То есть пока этот фрейм не трогаем

Напишем скрипт

```python
INPUT_PATH = '/home/user/PycharmProjects/pythonProject2/files/1_json_meta/*'
OUTPUT_PATH = '/home/user/PycharmProjects/pythonProject2/files/2_pandas_df'
```

```python
def convert_to_pandas(json_path):
    df_list = []
    name_list = []
    path_list = glob.glob(json_path)
    for i in path_list:
        with open(rf'{i}') as f:
            data = json.load(f)
        df = pd.DataFrame(data['streams'])
        df['FILE_NAME'] = str(os.path.basename(i))
        df_list.append(df)
    return df_list
    
    
```

```python
df_list = convert_to_pandas(INPUT_PATH)
```

```python
df = pd.concat(df_list)
```
