```python
import pandas as pd

df = pd.read_csv('F:\МФТИ\Хакатон\Получение датафрейма метаданных из видео\Metadata_dataframe_short_videos.csv')
df
```

```python
print(df.iloc[0])
```

Собираем датафрейм из имен и разрешений видеофайлов (индекс=0)

```python
df_resolution = df.loc[df['index']==0][['index', 'FILE_NAME', 'width', 'height', 'display_aspect_ratio']]
df_resolution
```

Переопределяем индекс, чтобы потом удалить строки по новому индексу

```python
df_resolution = df_resolution.rename(index=dict(zip(df_resolution.index, range(2837))))
df_resolution
```

Создаем список файлов с разрешением 16:9 (9:5)

```python
aspect_ratio = []
indexes_aspect = []
for i, row in df_resolution.iterrows():
    if row['display_aspect_ratio'] == '16:9' or row['display_aspect_ratio'] == '9:5':
        aspect_ratio.append(row['FILE_NAME'])
        indexes_aspect.append(i)
print(len(aspect_ratio))
print(aspect_ratio)
```

Удаляем файлы 16:9 и 9:5 из датафрейма

```python
df_resolution.drop(index=indexes_aspect, inplace=True)
df_resolution
```

```python
# переопределяем индекс
df_resolution = df_resolution.rename(index=dict(zip(df_resolution.index, range(2474))))
df_resolution
```

В датафрейме остались файлы, которые либо не подходят под 16:9, либо подходят, но имеют NaN-значение в display_aspect_ratio

Удаляем файлы с длиной и шириной, которые подходят под разрешение 16:9.
Возможно, позже дополнить условие, чтобы охватить диапазон около 16:9.

```python
calculated_ratio = []
indexes_calculated = []
for i, row in df_resolution.iterrows():
    if int(row['width'])%16 == 0 and int(row['height']%9) == 0:
        calculated_ratio.append(row['FILE_NAME'])
        indexes_calculated.append(i)
print(len(calculated_ratio))
print(calculated_ratio)
```

Проверяем еще раз по точным разрешениям.

Находим максимальные длину и ширину кадра, чтобы ограничить список условий.

```python
print(df_resolution.width.max())
print(df_resolution.height.max())
print(df_resolution.width.min())
print(df_resolution.height.min())
```

```python
list_other = []
indexes_other = []
for i, row in df_resolution.iterrows():
    if row['width'] == '640' and row['height'] == 360\
    or\
    row['width'] == '720' and row['height'] == 400\
    or\
    row['width'] == '854' and row['height'] == 480\
    or\
    row['width'] == '960' and row['height'] == 540:
        list_other.append(row['FILE_NAME'])
        indexes_other.append(i)
print(len(list_other))
print(list_other)
```

Больше нет файлов, подходящих под разрешение 16:9. 
Они находятся в списках aspect_ratio и calculated_ratio, их индексы - indexes_aspect и indexes_calculated.

Повторяем удаление строк датафрейма и переопределение индекса

```python
df_resolution.drop(index=indexes_calculated, inplace=True)
display(df_resolution)
```

```python
df_resolution = df_resolution.rename(index=dict(zip(df_resolution.index, range(2350))))
df_resolution
```

Сформируем общий список list_16_9

```python
list_16_9 = aspect_ratio + calculated_ratio
print(len(list_16_9))
print(len(df_resolution['width']))
```
