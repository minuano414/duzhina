Собираем датафрейм из имен и разрешений видеофайлов (индекс=0)

```python
df_resolution = df.loc[df['index']==0][['index', 'FILE_NAME', 'width', 'height', 'display_aspect_ratio']]
df_resolution
```

Переопределяем индекс, чтобы потом удалить строки по новому индексу

```python
df_resolution = df_resolution.rename(index=dict(zip(df_resolution.index, range(2837))))
df_resolution
```

display_aspect_ratio показывает соотношение ширины к высоте, т.е. 16:9 здесь означает отношение ширины к высоте.
Нам необходимо отношение высоты к ширине.

```python
df_resolution['display_aspect_ratio'].value_counts()
```

Все файлы имеют ширину больше высоты.

```python
df_resolution.width.hist()
```

```python
df_resolution.height.hist()
```

Соотношение сторон экрана 9:16 равно 0,5625.
Исследуем отклонение имеющихся значений соотношения сторон от этой цифры.

```python
df_resolution['screen_ratio'] = df_resolution['width'] / df_resolution['height']
df_resolution['ratio_var'] = df_resolution['screen_ratio'] - 9/16
df_resolution
```

```python
df_resolution.ratio_var.hist()
```

```python
a = df_resolution[df_resolution['ratio_var']<=0.9].shape[0]
b = df_resolution.loc[(df_resolution['ratio_var']>0.9) & (df_resolution['ratio_var']<=1.3)].shape[0]
c = df_resolution.loc[(df_resolution['ratio_var']>1.3) & (df_resolution['ratio_var']<=1.9)].shape[0]
display(a)
display(b)
display(c)
```

Из 2837 файлов 155 имеют примерную ширину 4 и высоту 3, 796 файлов имеют ширину 16 и высоту 9, 1886 файлов имеют ширину 12 и высоту 5.
