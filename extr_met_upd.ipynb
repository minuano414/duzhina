```python
#pip install ffmpeg-python
!pip install ffprobe-python
import ffmpeg
import ffprobe
import os.path
from pprint import pprint
import json
```

```python
def vid_metadata(path_f):
    try:
        with open('/home/tatiana_unix/metadata.json', 'w', encoding='utf-8') as f:
            for file in os.listdir(path_f):
                json.dump(ffmpeg.probe(f'{path_f}/{file}'), f, ensure_ascii=False, indent=4, separators=(',', ': '), sort_keys=True)
                print(f'\n[+] Метаданные файла: {file} успешно загружены.\n')
        #pprint(ffmpeg.probe(patn_f))
    except ffmpeg._run.Error:
        print('[-] Неподдерживаемый формат')
```

```python
vid_metadata('/home/tatiana_unix/shot_videos')
```
