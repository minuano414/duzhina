```python
import glob
import os 
import pandas as pd
import json
```

```python
INPUT_PATH = r'/home/user/PycharmProjects/pythonProject2/files/shot_videos/*'
OUTPUT_PATH = r'/home/user/PycharmProjects/pythonProject2/files/json_meta'
```

```python
path_list = glob.glob(INPUT_PATH)
```

```python
def meta_extr(path_file, json_file):
    '''Функция для извлечения метаданных из видео
     path_file - абслютный путь к папке с видеофайлами
     json_file - абсолютный путь выходных json файлов'''
    path_list = glob.glob(path_file)
    for i in path_list:
        file_name = str(os.path.basename(i))
        os.system(rf"ffprobe  -v quiet -print_format json -show_format -show_streams {i} > {OUTPUT_PATH}/{json_file}.json")
        print(i, file_name)
```

```python
meta_extr(INPUT_PATH)
```
